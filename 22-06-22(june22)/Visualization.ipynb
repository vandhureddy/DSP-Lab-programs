```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("Line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
xpoints=np.array([1,6])
ypoints=np.array([1,250])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
plt.plot(xpoints,ypoints,"--")
plt.show()
```

```python
import matplotlib.pyplot as plt
country=['USA','india','germany','uk','france']
gdp=[34000,45000,65000,22000,43000]
plt.bar(country,gdp)
plt.title('country vs gdp')
plt.xlabel('country')
plt.ylabel('gdp')
plt.show()
```
