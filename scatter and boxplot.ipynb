```python
#marker
import matplotlib.pyplot as plt
x=[0,1,2,3,4,5]
y=[3,4,5,6,7,8]
plt.scatter(x,y,marker='*')
```

```python
#color
import matplotlib.pyplot as plt
x=[8,2,9,5,4,5,8,2]
y=[3,4,5,6,7,8,4,7]
plt.scatter(x,y,marker='*',c='green')
```

```python
#size
import matplotlib.pyplot as plt
x=[8,2,9,5,4,5,8,2]
y=[3,4,5,6,7,8,4,7]
plt.scatter(x,y,marker='*',c='green',s=100)
```

```python
#line width
import matplotlib.pyplot as plt
x=[8,2,9,5,4,5,8,2]
y=[3,4,5,6,7,8,4,7]
plt.scatter(x,y,marker='*',c='green',s=100,linewidth=3)
```

```python
#edge color
import matplotlib.pyplot as plt
x=[8,2,9,5,4,5,8,2]
y=[3,4,5,6,7,8,4,7]
plt.scatter(x,y,marker='*',c='white',s=100,linewidth=3,edgecolor='black')
```

```python
#transparency of color-alpha 
import matplotlib.pyplot as plt
x=[8,2,9,5,4,5,8,2,3,8,1,6]
y=[3,4,5,6,7,8,4,7,4,8,2,3]
plt.scatter(x,y,marker='o',c='green',s=500,linewidth=3,alpha=0.2)
```

```python
#different colors
import matplotlib.pyplot as plt
x=[0,1,2,3,4,5,8,2,9,5,4,5,8,2,3]
y=[0,1,2,1,4,5,3,4,5,6,7,8,4,7,4]
color=["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta","pink","black"]
plt.scatter(x,y,c=color)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.rand(40)
y=np.random.rand(40)
z=np.random.rand(40)
colors=np.random.rand(40)
plt.scatter(x,y,s=z*1000,c=colors,alpha=0.2)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.randint(100,size=(100))
y=np.random.randint(100,size=(100))
colors=np.random.randint(100,size=(100))
sizes=10 * np.random.randint(100,size=(100))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5)
```
