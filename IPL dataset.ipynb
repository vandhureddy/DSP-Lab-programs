```python
#Grouping and Aggregating
#What is the average SOLD PRICE for each age category?
a.groupby('AGE')['SOLD PRICE'].mean()
```

```python
b=a.groupby('AGE')['SOLD PRICE'].mean().reset_index()
b
```

```python
#Average SOLD PRICE for Different Playing Roles in Each Age Category?
c=a.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
c
```

```python
#Joining dataframes
#Compare the average auction price for different ages and playing roles
comparison=c.merge(b,on='AGE',how='outer')
comparison
```

```python
#Re-naming columns
comparison.rename(columns={'SOLD PRICE_x':'sold price age role','SOLD PRICE_y':'sold price age'},inplace=True)
comparison.head(5)
```

```python
#Applying Operations to multiple columns
#Percentage change in SOLD PRICE
comparison['change']=comparison.apply(lambda rec:(rec.sold price age role-rec.sold price age)/rec.sold price age, axis = 1)
comparison
```

```python
#Filtering Records from Dataframe based on conditions
#Which players have hit more then 80 sixes in the IPL tournament so far?
a[a['SIXERS']>80][['PLAYER NAME','SIXERS']]
```
