```python
import matplotlib.pyplot as plt
import numpy as np
plt.title ("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
xpoints=np.array([0,2])
ypoints=np.array([0,100])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
plt.plot([5,200],[5,200],"--")
plt.show()
```
