```python
import matplotlib.pyplot as plt
country=['USA','INDIA','GERMANY','UK','FRANCE']
GDP_Per_Capita=[48000,38000,50000,45000,44000]
plt.bar(country,GDP_Per_Capita)
plt.title('country vs GDP Per Capita')
plt.xlabel('country')
plt.ylabel('GDP Per Capita')
plt.show()
```

```python
import matplotlib.pyplot as plt
country=['USA','INDIA','GERMANY','UK','FRANCE']
GDP_Per_Capita=[48000,38000,50000,45000,44000]
plt.barh(country,GDP_Per_Capita)
plt.title('country vs GDP Per Capita')
plt.xlabel('country')
plt.ylabel('GDP Per Capita')
plt.show()
```

```python
import matplotlib.pyplot as plt
country=['USA','INDIA','GERMANY','UK','FRANCE']
GDP_Per_Capita=[48000,38000,50000,45000,44000]
plt.bar(country,GDP_Per_Capita,color="pink")
plt.title('country vs GDP Per Capita')
plt.xlabel('country')
plt.ylabel('GDP Per Capita')
plt.show()
```
