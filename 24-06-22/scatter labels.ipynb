```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('cs.csv')
plt.scatter(data['AGE'],data['SALARY'],c='blue')
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.show()
x=data['AGE'].values.reshape(-1,1)
y=data['SALARY'].values
```

```python
reg=LinearRegression()
reg.fit(x,y)
print(reg.score(x,y))
print(reg.coef_)
print(reg.intercept_)
y_pred=reg.predict(x)
plt.scatter(data['NAME'],data['SALARY'],c='red')
plt.plot(data['NAME'],y_pred,c='blue',linewidth=2)
plt.xlabel("NAME")
plt.ylabel("SALARY")
plt.show()
print(reg.score(x,y_pred))
```
