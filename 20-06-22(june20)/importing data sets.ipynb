```python
#program for getting URL from the html page/website
page='WELCOME<a href="htttps://www.yahoo.com">HELLO HOW R U'
print(page)
def getURL(page):
    start_link=page.find("<a href")
    start_quote=page.find('"',start_link)
    end_quote=page.find('"',start_quote+1)
    url=page[start_quote+1:end_quote]
    #print(url)
    #getURL(page)
    return url
url=getURL(page)
print(url)
```

### b) Cleaning and Preparing the Data: Identify and Handle Missing Values, Data Formatting, Data Normalization Sets, Binning, Indicator variables

```python
#handling missing values
import numpy as np
import pandas as pd
a=pd.DataFrame({"Name":["Vanaja","navya","asha","anitha","pavani",np.nan,"HOPE",np.nan,"TAE","KOOK"],
"Id":[np.nan,390,232,313,181,315,np.nan,np.nan,242,847],
"Marks":[90,76,96,98,np.nan,94,84,np.nan,74,87],
"Country":["sk","tl","i",np.nan,"a","sl","c",np.nan,"ja","va"]})
a
```

```python
#removing all null values
a.dropna()
```

```python
#removing specified row
a.drop([1],axis=0)
```

```python
#removing specified column
a.dropna(axis=1)
```

```python
#replacing one value with another
a.replace(to_replace=[313],value=0)
```

```python
#replacing nan values with other values
a.replace(np.nan,0)
```

```python
#dropna using how
a.dropna(how="all")
```

```python
#dropna using how
a.dropna(how="any")
```

```python
a.replace(np.nan,"v")
```

```python
#dropna using subset
a.dropna(subset=['Id','Marks'])
```

### Data Normalization

```python
a2=a.copy()
column='Id'
a2["Id"]=(a2["Id"]/a2["Id"].abs().max())
print(a2)
```

```python
a3=a.copy()
column='Id'
a3["Id"]=(a3["Id"]-a3["Id"].min())/(a3["Id"].max()-a3["Id"].min())
display(a3)
```

```python
a4=a.copy()
column='Id'
a4["Id"]=(a4["Id"]-a4["Id"].mean())/a4["Id"].std()
display(a4)
```

```python
#converting datatype
a["Id"]=a["Id"].astype('float')
a
```

```python
#dropna using inplace
a4.dropna(inplace=True)
a4
```
